```python
import torch
import numpy as np
from torch.utils.data import Dataset, DataLoader
import os

class PoseDataset(Dataset):
    def __init__(self, base_dir, n_step=3, sequence_length=20):
        """
        :param base_dir: 비디오들이 있는 기본 폴더 경로
        :param n_step: 예측할 단계 수 (default: 3)
        :param sequence_length: 랜덤으로 뽑을 연속된 프레임의 길이 (default: 20)
        """
        # 모든 비디오 폴더에서 X3D.npy 파일을 찾아서 하나의 데이터셋으로 합침
        self.mesh_list = []
        
        # 모든 비디오 폴더를 탐색
        for video_folder in os.listdir(base_dir):
            video_path = os.path.join(base_dir, video_folder)
            if os.path.isdir(video_path):
                x3d_path = os.path.join(video_path, 'X3D.npy')
                if os.path.exists(x3d_path):
                    self.mesh_list.append(np.load(x3d_path))  # (num_frames, num_joints, 3)

        # 데이터셋이 비어있는지 확인
        if len(self.mesh_list) == 0:
            raise FileNotFoundError("X3D.npy 파일을 찾을 수 없습니다.")
        
        # 여러 비디오 데이터 합치기
        self.mesh = np.concatenate(self.mesh_list, axis=0)  # (total_num_frames, num_joints, 3)
        self.num_frames = self.mesh.shape[0]
        self.n_step = n_step
        self.sequence_length = sequence_length

    def __len__(self):
        # 연속된 n_step 프레임을 샘플링할 수 있는 범위
        return self.num_frames - self.sequence_length

    def __getitem__(self, idx):
        # 랜덤한 연속된 20개 프레임을 샘플링
        start_frame = np.random.randint(0, self.num_frames - self.sequence_length)

        # 연속된 20개 프레임 샘플링
        input_vector = self.mesh[start_frame:start_frame + self.n_step]      # (n_step, 17, 3)
        target_vector = self.mesh[start_frame + 1:start_frame + self.n_step + 1] # (n_step, 17, 3)

        # NumPy 배열을 PyTorch Tensor로 변환
        input_vector = torch.tensor(input_vector, dtype=torch.float32)
        target_vector = torch.tensor(target_vector, dtype=torch.float32)

        return input_vector, target_vector

# 데이터셋 경로 설정
base_dir = r"D:\pose_prediction\results"  # 비디오 폴더들이 있는 디렉토리 경로

# PoseDataset 생성
dataset = PoseDataset(base_dir, n_step=3, sequence_length=20)  # 예시: 20개의 연속된 프레임으로 설정
dataloader = DataLoader(dataset, batch_size=4, shuffle=True)

# 데이터 확인
for input_vector, target_vector in dataloader:
    if input_vector is None or target_vector is None:
        continue  # 유효하지 않은 배치는 건너뜁니다
    print("Input Shape:", input_vector.shape)   # (batch_size, n_step, 17, 3)
    print("Target Shape:", target_vector.shape) # (batch_size, n_step, 17, 3)
    break
```

```python
import torch
import numpy as np
from torch.utils.data import Dataset, DataLoader
import os
import cv2
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

class PoseDataset(Dataset):
    def __init__(self, base_dir, video_dir, n_step=3, sequence_length=20):
        """
        :param base_dir: results 폴더 (X3D.npy 파일이 있는 폴더)
        :param video_dir: video_split 폴더 (비디오 파일이 있는 폴더)
        :param n_step: 예측할 단계 수 (default: 3)
        :param sequence_length: 랜덤으로 뽑을 연속된 프레임의 길이 (default: 20)
        """
        self.base_dir = base_dir
        self.video_dir = video_dir
        self.n_step = n_step
        self.sequence_length = sequence_length

        # results 폴더에서 X3D.npy가 있는 비디오 폴더 리스트 가져오기
        self.video_list = [
            f for f in os.listdir(base_dir)
            if os.path.exists(os.path.join(base_dir, f, 'X3D.npy'))
        ]

        if not self.video_list:
            raise FileNotFoundError("X3D.npy 파일을 찾을 수 없습니다.")

    def __len__(self):
        """비디오 개수 기준으로 데이터셋 길이 설정"""
        return len(self.video_list)

    def __getitem__(self, idx):
        """해당 비디오의 npy 파일을 불러와서 랜덤한 구간을 선택"""
        video_name = self.video_list[idx]
        x3d_path = os.path.join(self.base_dir, video_name, 'X3D.npy')
        video_path = os.path.join(self.video_dir, f"{video_name}.mp4")

        # X3D.npy 불러오기
        mesh_data = np.load(x3d_path)  # (num_frames, num_joints, 3)
        num_frames = mesh_data.shape[0]

        # 랜덤한 시작 프레임 선택 (프레임 개수를 초과하지 않도록)
        start_frame = np.random.randint(0, max(1, num_frames - self.sequence_length))

        input_vector = mesh_data[start_frame:start_frame + self.n_step]  # (n_step, 17, 3)
        target_vector = mesh_data[start_frame + 1:start_frame + self.n_step + 1]  # (n_step, 17, 3)

        # NumPy 배열을 PyTorch Tensor로 변환
        input_vector = torch.tensor(input_vector, dtype=torch.float32)
        target_vector = torch.tensor(target_vector, dtype=torch.float32)

        # 시각화 실행
        self.visualize_pose(input_vector[0].numpy(), video_path, start_frame)

        return input_vector, target_vector

    def visualize_pose(self, pose, video_path, frame_idx):
        """
        3D 포즈 데이터를 시각화하고, 해당하는 비디오 프레임을 함께 보여줌
        :param pose: (17, 3) 크기의 포즈 데이터 (numpy)
        :param video_path: 비디오 파일 경로
        :param frame_idx: 현재 프레임 인덱스
        """
        fig = plt.figure(figsize=(10, 5))

        # 3D 포즈 시각화
        ax1 = fig.add_subplot(1, 2, 1, projection='3d')
        ax1.scatter(pose[:, 0], pose[:, 1], pose[:, 2], c='r', marker='o')
        ax1.set_xlabel("X")
        ax1.set_ylabel("Y")
        ax1.set_zlabel("Z")
        ax1.set_title(f"Pose, Frame {frame_idx}")

        # 비디오 프레임 표시
        if os.path.exists(video_path):
            cap = cv2.VideoCapture(video_path)
            cap.set(cv2.CAP_PROP_POS_FRAMES, frame_idx)
            ret, frame = cap.read()
            cap.release()

            if ret:
                frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
                ax2 = fig.add_subplot(1, 2, 2)
                ax2.imshow(frame)
                ax2.set_title(f"Video Frame: {frame_idx}")
                ax2.axis("off")
            else:
                print(f"⚠️ 프레임 {frame_idx}을(를) 불러올 수 없습니다.")

        plt.show()


# 데이터셋 경로 설정
base_dir = r"D:\pose_prediction\results"  # results 폴더
video_dir = r"D:\pose_prediction\video_split"  # video_split 폴더

# PoseDataset 생성
dataset = PoseDataset(base_dir, video_dir, n_step=3, sequence_length=20)
dataloader = DataLoader(dataset, batch_size=4, shuffle=True)

# 데이터 확인
for input_vector, target_vector in dataloader:
    if input_vector is None or target_vector is None:
        continue
    print("Input Shape:", input_vector.shape)   # (batch_size, n_step, 17, 3)
    print("Target Shape:", target_vector.shape) # (batch_size, n_step, 17, 3)
    break
```

```python
import torch
import torch.nn as nn

class PoseMLP(nn.Module):
    def __init__(self, input_dim=17*3, hidden_dim=128, output_dim=17*3):
        super(PoseMLP, self).__init__()
        self.model = nn.Sequential(
            nn.Linear(input_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, output_dim)
        )

    def forward(self, x):
        x = x.view(x.shape[0], -1)  # (batch_size, 17*3)
        x = self.model(x)
        x = x.view(x.shape[0], 17, 3)  # (batch_size, 17, 3)
        return x

# 모델 생성
model = PoseMLP()
print(model)

# 테스트 입력
dummy_input = torch.randn(32, 17, 3)  # (batch_size, num_joints, 3)
output = model(dummy_input)
print("Output shape:", output.shape)  # 예상: (32, 17, 3)
```

```python
import torch.optim as optim

# 손실 함수 및 옵티마이저 정의
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# 학습 루프
num_epochs = 25  # 원하는 에포크 수 설정
for epoch in range(num_epochs):
    total_loss = 0.0

    for input_vector, target_vector in dataloader:
        optimizer.zero_grad()  # 기울기 초기화

        outputs = model(input_vector)  # 예측값
        loss = criterion(outputs, target_vector)  # MSE 손실 계산

        loss.backward()  # 역전파
        optimizer.step()  # 가중치 업데이트

        total_loss += loss.item()

    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {total_loss/len(dataloader):.6f}")
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 데이터 로딩
input_vector, target_vector = next(iter(dataloader))  # 첫 번째 배치 가져오기
outputs = model(input_vector)  # 모델 예측

# 첫 번째 샘플 선택 (batch_size 중 하나만 시각화)
target_sample = target_vector[0].detach().numpy()  # (17, 3)
output_sample = outputs[0].detach().numpy()  # (17, 3)

# 관절 연결 순서 (limbSeq)
limbSeq = [[0, 1], [1, 2], [1, 5], [1, 8], [2, 3], [3, 4], [5, 6], [6, 7],
           [8, 9], [8, 13], [9, 10], [10, 11], [11, 12], [13, 14], [14, 15], [15, 16]]

# 3D 플롯 설정
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111, projection='3d')

# 타겟 데이터(파란색)와 예측 데이터(빨간색) 각각 플롯
for joint1, joint2 in limbSeq:
    # Target (ground truth)
    x_t, y_t, z_t = target_sample[joint1]
    x_t2, y_t2, z_t2 = target_sample[joint2]
    ax.plot([x_t, x_t2], [y_t, y_t2], [z_t, z_t2], 'b-', linewidth=2, label="Target" if joint1 == 0 else "")

    # Output (prediction)
    x_o, y_o, z_o = output_sample[joint1]
    x_o2, y_o2, z_o2 = output_sample[joint2]
    ax.plot([x_o, x_o2], [y_o, y_o2], [z_o, z_o2], 'r-', linewidth=2, label="Output" if joint1 == 0 else "")

# 관절 점 플롯 (파란색: 타겟, 빨간색: 출력)
ax.scatter(target_sample[:, 0], target_sample[:, 1], target_sample[:, 2], color='blue', label="Target Joints", s=30)
ax.scatter(output_sample[:, 0], output_sample[:, 1], output_sample[:, 2], color='red', label="Output Joints", s=30)

# 그래프 설정
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("3D Pose Prediction vs Target")
ax.legend()

plt.show()
```
