```python
import os
import cv2

def split_video(input_video_path, output_dir, frames_per_chunk=1000):
    # ë¹„ë””ì˜¤ ì½ê¸°
    cap = cv2.VideoCapture(input_video_path)
    
    if not cap.isOpened():
        print(f"Error: ë¹„ë””ì˜¤ë¥¼ ì—´ ìˆ˜ ì—†ìŠµë‹ˆë‹¤. ({input_video_path})")
        return

    # ë¹„ë””ì˜¤ ê¸°ë³¸ ì •ë³´ ê°€ì ¸ì˜¤ê¸°
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    fps = cap.get(cv2.CAP_PROP_FPS)
    video_name = os.path.basename(input_video_path).split('.')[0]

    print(f"{input_video_path}: ì´ {total_frames}í”„ë ˆì„ì´ ìˆìŠµë‹ˆë‹¤.")

    # ì €ì¥í•  ë””ë ‰í† ë¦¬ ì¤€ë¹„
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    frame_count = 0
    chunk_index = 0
    frames = []

    while True:
        ret, frame = cap.read()
        if not ret:
            break
        
        frames.append(frame)
        frame_count += 1

        # 1000í”„ë ˆì„ë§ˆë‹¤ ë¹„ë””ì˜¤ë¡œ ì €ì¥
        if frame_count % frames_per_chunk == 0 or frame_count == total_frames:
            chunk_index += 1
            chunk_output_path = os.path.join(output_dir, f"{video_name}_chunk{chunk_index}.mp4")
            
            # 1000í”„ë ˆì„ì”© ì €ì¥
            fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # MP4 codec
            out = cv2.VideoWriter(chunk_output_path, fourcc, fps, (frame.shape[1], frame.shape[0]))

            for f in frames:
                out.write(f)
            out.release()  # ë¹„ë””ì˜¤ íŒŒì¼ ì €ì¥ í›„ ë¦¬ì†ŒìŠ¤ í•´ì œ

            print(f"'{chunk_output_path}' ì €ì¥ ì™„ë£Œ.")
            frames = []  # ìƒˆë¡œ ì‹œì‘í•  í”„ë ˆì„ë“¤ ì´ˆê¸°í™”

    cap.release()
    print(f"{input_video_path}: ë¹„ë””ì˜¤ ë¶„í•  ì™„ë£Œ.")

def split_all_videos_in_directory(video_dir, output_dir, frames_per_chunk=1000):
    video_extensions = ('.mp4', '.avi', '.mov', '.mkv')  # ì§€ì›í•˜ëŠ” ë¹„ë””ì˜¤ í™•ì¥ì
    video_files = [f for f in os.listdir(video_dir) if f.endswith(video_extensions)]
    
    if not video_files:
        print("âš ï¸ ì²˜ë¦¬í•  ë¹„ë””ì˜¤ íŒŒì¼ì´ ì—†ìŠµë‹ˆë‹¤.")
        return

    print(f"ğŸ¥ ì´ {len(video_files)}ê°œì˜ ë¹„ë””ì˜¤ë¥¼ ì²˜ë¦¬í•©ë‹ˆë‹¤.\n")

    for video in video_files:
        video_path = os.path.join(video_dir, video)
        print(f"â–¶ '{video}' ì²˜ë¦¬ ì‹œì‘...")
        split_video(video_path, output_dir, frames_per_chunk)
    
    print("ëª¨ë“  ë¹„ë””ì˜¤ ì²˜ë¦¬ ì™„ë£Œ.")

# ì‚¬ìš© ì˜ˆì‹œ
video_dir = './video'         # ë¹„ë””ì˜¤ê°€ ìˆëŠ” ë””ë ‰í† ë¦¬ ê²½ë¡œ
output_dir = './video_split'       # ë¶„í• ëœ ë¹„ë””ì˜¤ ì €ì¥ ê²½ë¡œ
split_all_videos_in_directory(video_dir, output_dir, frames_per_chunk=500)
```
